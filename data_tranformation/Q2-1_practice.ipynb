```python
import pandas as pd
```

```python
tournament_data= pd.read_csv('tournaments.csv')
goals_data= pd.read_csv('goals.csv')
```

```python
tournament_data=tournament_data.rename(columns={'winner': 'team_name'})
tournament_data.head()
```

```python
goals_data['team1']=goals_data['match_name'].apply(lambda x : x.split(' v ')[0])
goals_data['team2']=goals_data['match_name'].apply(lambda x : x.split(' v ')[1])
goals_data.head()
```

```python
goals_data.columns
```

```python
team1_first_goal=goals_data.groupby(['tournament_id', 'team_name', 'match_id']).min()['minute_regulation'].reset_index()
team1_first_goal=team1_first_goal.rename(columns={'team_name': 'team1', 'minute_regulation': 'team1_first_goal_minute'})
team1_first_goal
```

```python
goals_data=pd.merge(goals_data, team1_first_goal, on= ['tournament_id', 'match_id', 'team1' ], how='left')

goals_data
```

```python
team2_first_goal=team1_first_goal.rename(columns={'team1': 'team2', 'team1_first_goal_minute': 'team2_first_goal_minute'})
goals_data=pd.merge(goals_data, team2_first_goal, on=['tournament_id', 'match_id', 'team2' ], how='left')
goals_data
```

```python
goals_data.loc[ goals_data['team1_first_goal_minute'].isnull(), 'team1_first_goal_minute' ] = 200
goals_data.loc[ goals_data['team2_first_goal_minute'].isnull(), 'team2_first_goal_minute' ] = 200

#goals_data.head(15)
```

```python

goals_data['first_goal_country']=goals_data['team1']
goals_data.loc[(goals_data['team1_first_goal_minute'] > goals_data['team2_first_goal_minute']), 'first_goal_country']= goals_data.loc[(goals_data['team1_first_goal_minute'] >  goals_data['team2_first_goal_minute']), 'team2']
goals_data
```
