```python
import pandas as pd
import numpy as np
```

```python
tournament_data= pd.read_csv('tournaments.csv')
region_data=pd.read_csv('teams.csv')
```

```python

tournament_data=tournament_data[['tournament_id','host_country', 'winner', 'host_won' ]]
tournament_data=tournament_data.rename(columns={'winner' : 'team_name'})
tournament_data
```

```python
region_data[['team_name', 'region_name']]
```

```python
merged=pd.merge(tournament_data, region_data, on='team_name', how='inner')
merged=merged.rename(columns={'team_name' : 'winner'})
merged.sort_values('region_name')
```
