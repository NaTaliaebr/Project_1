```python
import pandas as pd
import numpy as np
```

```python
tournament_data= pd.read_csv('tournaments.csv')
region_data=pd.read_csv('teams.csv')
```

```python

tournament_data=tournament_data[['tournament_id','host_country', 'winner', 'host_won' ]]
tournament_data=tournament_data.rename(columns={'winner' : 'team_name'})
tournament_data
```

```python
region_data[['team_name', 'region_name']]
```

```python
merged=pd.merge(tournament_data, region_data, on='team_name', how='inner')
merged=merged.rename(columns={'team_name' : 'winner'})
merged.sort_values('region_name')
```

```python
merged[['tournament_id', 'host_country','host_won', 'winner', 'region_name'  ]]

```

```python
#end of data transformation code && start of visualization code
import matplotlib.pyplot as plt
```

```python
tournaments = merged['tournament_id'].values
tournaments
regions = merged['region_name'].values
regions
```

```python
 #find count per region
region_count = merged["region_name"].value_counts()




 #Use DataFrame.plot() in order to create a bar chart of the data
region_count.plot(kind="bar",color = "b",
                   xlabel="Region of Host Winner",
                   ylabel="Frequency")

plt.show()
region_count
```

```python

region_count.plot(kind = "pie" , autopct="%1.0f%%")
plt.title( "Regions of Host Winners")
plt.ylabel('')
plt.show()
```
