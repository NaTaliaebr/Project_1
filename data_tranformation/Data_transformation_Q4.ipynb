```python
#Q4: To determine the range of penalty kicked, awarded and converted (scored) of the winners
```

```python
import pandas as pd
import numpy as np
```

```python
tournament_data= pd.read_csv('tournaments.csv')
penalty_data=pd.read_csv('penalty_kicks.csv')
```

```python
penalty_data
```

```python
tournament_data=tournament_data[['tournament_id', 'winner']]
tournament_data=tournament_data.rename(columns={'winner': 'team_name'})
tournament_data
```

```python
penalty_data=penalty_data[['penalty_kick_id', 'tournament_id', 'team_name', 'converted', 'match_id']]
penalty_data
```

```python
grouped=penalty_data.groupby(['team_name']).agg({'penalty_kick_id': 'count', 'converted' : 'sum'}).reset_index()
grouped=grouped.rename(columns={'penalty_kick_id': 'total_penalty_kicks'})
grouped['conversion_percentage(%)']=grouped['converted']/grouped['total_penalty_kicks']*100
grouped.head()
```

```python
final=pd.merge(tournament_data, grouped, on='team_name', how='left')
final=final[['team_name', 'total_penalty_kicks', 'converted', 'conversion_percentage(%)']].drop_duplicates()
final
```
