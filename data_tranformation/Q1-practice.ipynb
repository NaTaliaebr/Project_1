```python
import pandas as pd
```

```python
tournament_data= pd.read_csv('tournaments.csv')
bookings_data= pd.read_csv('bookings.csv')
```

```python
#bookings_data=bookings_data[['tournament_id', 'team_name', 'booking_id', 'minute_regulation','yellow_card', 'red_card' , 'second_yellow_card']].head(2)
bookings_data
```

```python
#tournament_data=tournament_data[['tournament_id', 'winner']].head(2)
tournament_data=tournament_data.rename(columns={'winner': 'team_name'})
tournament_data
```

```python
winner_booking_info=pd.merge(tournament_data, bookings_data, on= ['tournament_id', 'team_name'])
winner_booking_info[['tournament_id','team_name', 'yellow_card', 'red_card']]
winner_booking_info.groupby(['tournament_id','team_name']).agg({'yellow_card': 'sum', 'red_card': 'sum'}).reset_index()
```

```python
non_winner_booking_info=pd.merge(tournament_data[ [ 'tournament_id', 'team_name' ] ], \
                                 bookings_data, how='outer', on=['tournament_id', 'team_name'], \
                                 indicator=True)

non_winner_booking_info=non_winner_booking_info.loc[non_winner_booking_info['_merge']== 'right_only']
non_winner_booking_info[['tournament_id','team_name', 'yellow_card', 'red_card']]
non_winner_booking_info.groupby(['tournament_id','team_name']).agg({'yellow_card': 'sum', 'red_card': 'sum'}).reset_index()

```
